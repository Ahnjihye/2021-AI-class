## pandas를 활용하여 데이터를 가공할 수 있는 형태로 변경하기 

````
import pandas as pd

data = pd.read_csv('iris (1).csv', encoding='cp949')
data
````

# 2. 그래프 시각화

### 2-1. 꽃받침길이(x축), 꽃받침너비(y축)으로 산점도 그리기

````
import matplotlib.pyplot as plt

plt.scatter(data['꽃받침길이'], data['꽃받침너비'])
plt.show()
````

### 2-2. setosa의 꽃받침길이(x축), 꽃받침너비(y축)으로 산점도 그리기

````
setosa = data['붓꽃품종']=='setosa'
setosa
````

````
setosa = data[data['붓꽃품종']=='setosa']
setosa
````

````
plt.scatter(setosa['꽃받침길이'], setosa['꽃받침너비'], color = 'indigo')
plt.show()
````

### 2-3. versicolor의 꽃받침길이(x축), 꽃받침너비(y축)으로 산점도 그리기

````
versi = data[data['붓꽃품종']=='versicolor']
versi
````

````
plt.scatter(versi['꽃받침길이'], versi['꽃받침너비'], color = 'crimson')
plt.show()
````

### 2-4. virginica의 꽃받침길이(x축), 꽃받침너비(y축)으로 산점도 그리기

````
vergi = data[data['붓꽃품종']=='virginica']
vergi
````

````
plt.scatter(vergi['꽃받침길이'], vergi['꽃받침너비'], color = 'limegreen')
plt.show()
````

#### 2-5. 세 개 품종의 꽃받침길이(x축), 꽃받침너비(y축)으로 산점도 그리기

````
plt.figure(dpi=100)
plt.scatter(setosa['꽃받침길이'], setosa['꽃받침너비'], color = 'indigo', label='setosa')
plt.scatter(versi['꽃받침길이'], versi['꽃받침너비'], color = 'crimson', label='versicolor')
plt.scatter(vergi['꽃받침길이'], vergi['꽃받침너비'], color = 'limegreen', label='verginica')
plt.legend()
plt.show()
````
